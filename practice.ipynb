```python
import matplotlib.pyplot as plt
import numpy as np

a = -1
b = 0
c = 0

# create 1000 equally spaced points between -10 and 10
x = np.linspace(-10, 10, 1000)

# calculate the y value for each element of the x vector
y = a*(x**2) + b*x + c  

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
ages_x_array = np.array(ages_x).reshape(-1, 1)
dev_y_array = np.array(dev_y).reshape(-1, 1)

plt.scatter(ages_x, dev_y, 1)

plt.xlabel("Age")
plt.ylabel("Salary")

model = LinearRegression().fit(ages_x_array, dev_y_array)
line = model.predict(ages_x_array)

print("Coefficients: \n", model.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(dev_y_array, line))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(dev_y_array, line))
print("Pearson Correlation Coefficient: %.2f" % r_regression(dev_y_array, line))

plt.plot(ages_x_array, line)

plt.legend()

plt.show()
```
